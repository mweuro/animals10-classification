```python
from PIL import Image
import torch
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
```

```python
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])
```

```python
def load_single_image(image_path):
    image = Image.open(image_path).convert('RGB')
    image_tensor = transform(image)
    return image_tensor
```

```python
image_tensor = load_single_image('data/raw-img/gatto/1.jpeg')
print(f"Shape: {image_tensor.shape}")  # torch.Size([3, 224, 224])
```

```python
import matplotlib.pyplot as plt
import numpy as np

def show_tensor_image(tensor):
    """
    Wywietla tensor obrazu o shape (3, 224, 224)
    """
    # Konwersja tensor -> numpy i zmiana kolejnoci wymiarw
    image = tensor.numpy().transpose(1, 2, 0)
    
    # Jeli tensor by znormalizowany, odwr normalizacj
    image = np.clip(image, 0, 1)
    
    plt.figure(figsize=(8, 8))
    plt.imshow(image)
    plt.axis('off')
    plt.show()

# Przykad uycia
show_tensor_image(image_tensor)
```
